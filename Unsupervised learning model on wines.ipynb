# KMeans Clustering

```python
from sklearn.cluster import KMeans
# Finding optimal clusters using Elbow diagram
kpoints = range(2,10)
def draw_elbow(elbow_df):
    inertias = []
    for k in kpoints:
        inertias.append(KMeans(n_clusters=k, init = "k-means++", random_state=42).fit(elbow_df).inertia_)
    plt.figure(figsize = (15, 6))
    plt.plot(kpoints, inertias, "-o")
    plt.title("Number of Clusters vs Distance")
    plt.xlabel("Number of Clusters")
    plt.ylabel("Distance")
    plt.show()
```

```python
draw_elbow(df)
```

optimal k value is 4 

```python
kmeans = KMeans(n_clusters=4, init = "k-means++", random_state=42)
kmeans.fit(df)
k_clusters = kmeans.labels_
k_clusters
```

```python
result=df.copy()
```

```python
result['Clusters_kmeans']=k_clusters

result.head()
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1],c=k_clusters)
```

```python
#Using the dendrogram to find the optimal number of clusters
import scipy.cluster.hierarchy as sch
plt.subplots(figsize=(20,10))
dendrogram = sch.dendrogram(sch.linkage(df, method = "ward"))
plt.hlines(0.3,0,5000,colors="r", linestyles="solid") # for the horizontal cut line 
plt.title("dendrogram")

plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 4 , affinity = "euclidean" , linkage = "ward")

model_hc = hc.fit(df)

HC_clusters = model_hc.labels_
plt.scatter(df.iloc[:,0],df.iloc[:,1],c=HC_clusters)
```
